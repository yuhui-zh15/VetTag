```python
f1 = open('/home/yuhuiz/Transformer/data/csu/csu_metamap_test.tsv')
f2 = open('/home/yuhuiz/Transformer/data/csu/csu_test.tsv')
from collections import defaultdict
out = {}
for i in range(42):
    out[str(i)] = defaultdict(int)
for line in f1:
    splitline1 = line.strip().split()
    splitline2 = f2.readline().strip().split('\t')
    if len(splitline2) == 1: continue
    labels = splitline2[1].split()
    for label in labels:
        if label == '41': continue
        for word in splitline1:
            out[label][word] += 1
f1.close()
f2.close()
```
